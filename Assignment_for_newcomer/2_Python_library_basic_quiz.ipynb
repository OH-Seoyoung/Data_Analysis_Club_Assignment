> 1번. 4x2 크기의 정수로된 numpy array 를 만들고 아래의 두가지 특성과 array를 출력하세요.

[[12 32]

> 7번. 주어진 numpy array에 대하여, 행과 열 기준으로 최댓값과 최솟값을 각각 출력하세요.  

**실행 예 :**   
[[34 43 73]  
 [82 22 12]  
 [53 94 66]]  

[34 22 12]  
[34 12 53]  
[82 94 73]  
[73 82 94]  

```python
arr = np.array([[34,43,73],[82,22,12],[53,94,66]])
print(arr)
print()

# --- start code ---
print(np.amin(arr, 0))
print(np.amin(arr, 1))
print(np.amax(arr, 0))
print(np.amax(arr, 1))
# --- end code ---
```
